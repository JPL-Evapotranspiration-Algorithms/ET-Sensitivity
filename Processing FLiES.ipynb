```python
import numpy as np
import pandas as pd
from dateutil import parser
from FLiES_sensitivity import process_FLiES_table
```

```python
FLiES_inputs_df = pd.read_csv("calval_final_50_FLiES_inputs.csv")
FLiES_inputs_df = FLiES_inputs_df[['tower', 'lat', 'lon', 'orbit', 'scene', 'tile', 'time_UTC', 'date_UTC', 'doy', 'time_solar', "AOT", "COT", "vapor_gccm", "ozone_cm", "SZA", "KG"]]
FLiES_inputs_df = FLiES_inputs_df.dropna()
FLiES_inputs_df
```

```python
FLiES_inputs_df.date_UTC = FLiES_inputs_df.date_UTC.apply(lambda d: parser.parse(d).strftime("%Y-%m-%d"))
FLiES_inputs_df.date_UTC
```

```python
calval_final_50 = pd.read_csv("calval_final_50_by_column.csv")
calval_final_50
```

```python
calval_final_50.columns
```

```python
tower_locations = calval_final_50[["tower", "lat", "lon"]].groupby("tower").first().reset_index()
tower_locations.to_csv("tower_locations.csv", index=False)
tower_locations
```

```python
elevation = pd.read_csv("tower_elevation.csv")
elevation["elevation_km"] = elevation.elevation_m / 1000
elevation = elevation[["tower", "elevation_km"]]
elevation
```

```python
calval_final_50.date_UTC = calval_final_50.date_UTC.apply(lambda d: parser.parse(d).strftime("%Y-%m-%d"))
calval_final_50.date_UTC
```

```python
FLiES_inputs_df.columns
```

```python
df = pd.merge(FLiES_inputs_df, calval_final_50, how="inner", on=["tower", "date_UTC"], suffixes=["","_Y"])
df
```

```python
df.columns
```

```python
df = pd.merge(df, elevation, how="left", on="tower")
df
```

```python
df.to_csv("calval_final_50_FLiES_inputs_clean.csv", index=False)
```

```python
FLiES_outputs_df = process_FLiES_table(df)
FLiES_outputs_df.to_csv("calval_final_50_FLiES_outputs_baseline.csv")
FLiES_outputs_df
```
