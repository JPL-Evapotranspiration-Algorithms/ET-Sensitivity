```python
import numpy as np
import pandas as pd
from datetime import timedelta
from dateutil import parser
```

```python
source_df = pd.read_csv("calval_final_50.csv")
source_df.drop(columns=["Unnamed: 0"], inplace=True)
source_df.insert(5, "date_UTC", source_df.time_UTC.apply(lambda time_UTC: parser.parse(str(time_UTC)).date()), allow_duplicates=True)
source_df.insert(6, "time_solar", source_df.apply(lambda row: parser.parse(str(row.time_UTC)) + timedelta(hours=(np.radians(row.lon) / np.pi * 12)), axis=1), allow_duplicates=True)
source_df = source_df[["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar", "product", "variable", "med3x3"]]
source_df
```

```python
NDVI_albedo_df = source_df[source_df.apply(lambda row: "NDVI" in row.variable or "albedo" in row.variable, axis=1)]
NDVI_albedo_df = NDVI_albedo_df.pivot_table(index=["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar", "product"], columns="variable", values="med3x3").reset_index()
NDVI_albedo_df
```

```python
# instantaneous_df = source_df[source_df.apply(lambda row: "NDVI" not in row.variable and "albedo" not in row.variable, axis=1)]
instantaneous_df = source_df.copy()
instantaneous_df = instantaneous_df.pivot_table(index=["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar", "product"], columns="variable", values="med3x3").reset_index()
instantaneous_df
```

```python
metadata_df = instantaneous_df.groupby(by=["tower", "time_UTC"]).first().reset_index()[["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar", "product"]]
metadata_df
```

```python
df = pd.merge(left=metadata_df, right=NDVI_albedo_df, how="inner")
df = pd.merge(left=df, right=instantaneous_df, how="inner")
df = df[df.albedo != 0]
df.to_csv("calval_final_50_by_column.csv")
df
```
