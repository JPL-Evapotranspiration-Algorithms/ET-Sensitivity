```python
import pandas as pd
```

```python
filename = "combined_eco_flux.csv"
df_source = pd.read_csv(filename)
df_source.head()
```

```python
df_header = df_source[["ID", "time_utc"]]

df_header = df_header.rename(columns={
    "ID": "tower", 
    "time_utc": "time_UTC"
})

df_header
```

```python
df_body = df_source[[column for column in df_source.columns if column.endswith("_filt")]]
df_body = df_body.rename(columns={
    "LE_filt": "LE_insitu",
    "H_filt": "H_insitu",
    "NETRAD_filt": "Rn_insitu",
    "G_filt": "G_insitu"
})

df_body
```

```python
df = pd.concat([df_header, df_body], axis=1)
df
```

```python
df.to_csv("fluxnet.csv", index=False)
```
