```python
import os
import xarray as xr
import rioxarray
```

```python
# specify the directory containing the .tif files
directory = '/Users/gregoryhalverson/data/ECOSTRESS_example/ECOv002_L3T_SEB_21485_013_11SPS_20220420T211350_0700_01'

# get a list of all .tif files in the directory
tif_files = [f for f in os.listdir(directory) if f.endswith('.tif')]

# read each .tif file into a rioxarray dataset and store in a dictionary
# also extract the layer name from the filename
datasets = {}
for f in tif_files:
    layer_name = os.path.splitext(f)[0].split('_')[-1]  # get part after last underscore
    datasets[layer_name] = rioxarray.open_rasterio(os.path.join(directory, f))

# combine all datasets into one
xds = xr.Dataset(datasets)

xds
```
