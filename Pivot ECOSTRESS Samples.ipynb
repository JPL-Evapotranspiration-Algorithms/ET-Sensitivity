```python
import numpy as np
import pandas as pd
from datetime import timedelta
from dateutil import parser
```

```python
source_df = pd.read_csv("calval_final_50.csv")
source_df.drop(columns=["Unnamed: 0"], inplace=True)
source_df.insert(5, "date_UTC", source_df.time_UTC.apply(lambda time_UTC: parser.parse(str(time_UTC)).date()), allow_duplicates=True)
source_df.insert(6, "time_solar", source_df.apply(lambda row: parser.parse(str(row.time_UTC)) + timedelta(hours=(np.radians(row.lon) / np.pi * 12)), axis=1), allow_duplicates=True)
source_df = source_df[["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar", "product", "variable", "med3x3"]]
source_df
```

```python
source_df[source_df.variable == "NDVI"]
```

```python
source_df[source_df.variable == "Ta"]
```

```python
df = source_df.pivot_table(index=["tower", "lat", "lon", "orbit", "scene", "tile", "time_UTC", "date_UTC", "time_solar"], columns="variable", values="med3x3").reset_index()
df["year"] = df.time_solar.apply(lambda t: t.year)
df["Ta_C"] = df.Ta
df["ST_K"] = df.LST
df
```

```python
NDVI_albedo_df = df[["tower", "date_UTC", "NDVI", "albedo"]].dropna()
NDVI_albedo_df
```

```python
df = df.drop(["NDVI", "albedo"], axis=1)
df = pd.merge(df, NDVI_albedo_df, how="left", on=["tower", "date_UTC"])
df
```

```python
df["ST_C"] = df.ST_K - 273.15
df = df[~np.isnan(df.ST_C)]
df = df[df.ST_C > 0]
df
```

```python
df = df[~np.isnan(df.NDVI)]
df = df[df.NDVI > 0]
df = df[df.NDVI < 1]
df
```

```python
df = df[~np.isnan(df.albedo)]
df = df[df.albedo > 0]
df = df[df.albedo < 1]
df
```
