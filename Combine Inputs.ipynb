```python
import pandas as pd
import numpy as np
from dateutil import parser
from datetime import datetime
```

```python
flies_outputs_filename = "calval_final_50_PTJPL_inputs.csv"
flies_outputs = pd.read_csv(flies_outputs_filename)
flies_outputs.tower = flies_outputs.tower.apply(lambda x: str(x))
flies_outputs.time_UTC = flies_outputs.time_UTC_Y
flies_outputs = flies_outputs[flies_outputs.time_UTC.apply(lambda x: str(x) != "nan")]
flies_outputs.time_UTC = flies_outputs.time_UTC.apply(lambda x: str(datetime.strftime(parser.parse(str(x)), "%Y-%m-%d %H:%M:%S")))
flies_outputs
```

```python
fluxnet_filename = "fluxnet.csv"
fluxnet = pd.read_csv(fluxnet_filename)
fluxnet.tower = fluxnet.tower.apply(lambda x: str(x))
fluxnet = fluxnet[fluxnet.time_UTC.apply(lambda x: str(x) != "nan")]
fluxnet["date_UTC"] = fluxnet.time_UTC.apply(lambda x: str(datetime.strftime(parser.parse(str(x)), "%Y-%m-%d")))
fluxnet.time_UTC = fluxnet.time_UTC.apply(lambda x: str(datetime.strftime(parser.parse(str(x)), "%Y-%m-%d %H:%M:%S")))
fluxnet
```

```python
merged_df = pd.merge(flies_outputs, fluxnet, on=["tower", "date_UTC"], how="left")
merged_df
```

```python
radiation_df = pd.read_csv("radiation.csv")

radiation_df = radiation_df.rename(columns={
    "SWin": "SWin_insitu",
    "SWout": "SWout_insitu",
    "LWin": "LWin_insitu",
    "LWout": "LWout_insitu"
})

radiation_df
```

```python
merged_df = pd.merge(merged_df, radiation_df, on=["tower", "doy", "hour_of_day"], how="left")
merged_df
```

```python
merged_df.columns
```

```python
if "Ta" in merged_df and "Ta_C" not in merged_df:
    # input_df.rename({"Ta": "Ta_C"}, inplace=True)
    merged_df["Ta_C"] = merged_df["Ta"]

merged_df = merged_df[merged_df.fAPARmax.apply(lambda fAPARmax: fAPARmax > 0.001)]
merged_df = merged_df[merged_df.NDVI.apply(lambda NDVI: NDVI > 0.05)]
merged_df.columns
```

```python
if "time_UTC" not in merged_df and "time_UTC_Y" in merged_df:   
    merged_df.time_UTC = merged_df.time_UTC_Y

columns = list(merged_df.columns)
columns = [column for column in columns if not column.endswith("_x")]
columns = [column for column in columns if not column.endswith("_X")]
columns = [column for column in columns if not column.endswith("_y")]
columns = [column for column in columns if not column.endswith("_Y")]
columns = [column for column in columns if not column.startswith("Unnamed")]
columns
```

```python
merged_df = merged_df[columns]
merged_df
```

```python
merged_df = merged_df.drop_duplicates(subset=["tower", "date_UTC"])
merged_df
```

```python
merged_df.to_csv("inputs.csv", index=False)
```
