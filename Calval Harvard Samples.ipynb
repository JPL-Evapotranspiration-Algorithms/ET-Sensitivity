```python
import numpy as np
import pandas as pd
from datetime import timedelta
from dateutil import parser
```

```python
source_df = pd.read_csv("calval_harvard.csv")
source_df["time_solar"] = source_df.apply(lambda row: parser.parse(str(row.time_UTC)) + timedelta(hours=(np.radians(row.lon) / np.pi * 12)), axis=1)
source_df["date_UTC"] = source_df.time_UTC.apply(lambda time_UTC: parser.parse(str(time_UTC)).date())
source_df
```

```python
NDVI_albedo_df = source_df[source_df.apply(lambda row: "NDVI" in row.variable or "albedo" in row.variable, axis=1)]
NDVI_albedo_df = NDVI_albedo_df.pivot_table(index="date_UTC", columns="variable", values="med3x3").reset_index()
NDVI_albedo_df
```

```python
instantaneous_df = source_df[source_df.apply(lambda row: "NDVI" not in row.variable and "albedo" not in row.variable, axis=1)]
instantaneous_df = instantaneous_df.pivot_table(index="time_UTC", columns="variable", values="med3x3").reset_index().dropna()
instantaneous_df
```

```python
metadata_df = instantaneous_df.groupby(by=["tower", "time_UTC"]).first().reset_index()[["tower", "lat", "lon", "time_UTC", "date_UTC", "time_solar"]]
metadata_df
```

```python
df = pd.merge(left=metadata_df, right=NDVI_albedo_df, how="inner")
df = pd.merge(left=df, right=instantaneous_df, how="inner")
df = df[df.albedo != 0]
df.to_csv("calval_harvard_columns.csv")
df
```

```python
ha2 = df[df.tower == "US-Ha2"]
ha2.to_csv("ha2.csv")
ha2
```
